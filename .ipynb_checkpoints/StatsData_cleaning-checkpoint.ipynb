```python
full = pd.concat([seas07,seas08,seas09,seas10,seas11,seas12,seas13,seas14,seas15,seas16,seas17,seas18,seas19,seas20,seas21,seas22])
```

```python
full.dropna(inplace = True)
```

```python
full['Age'] = full['Age'].astype(int)
full['Season'] = full['Season'].astype(int)
```

```python
def season_weights(decay_factor, num_seasons):
    weights = []
    if num_seasons == 1:
        weights = [1]
    else:
        weight_sum = sum([decay_factor ** i for i in range(num_seasons)])
        player_weights = [(decay_factor ** i) / weight_sum for i in range(num_seasons)]
        weights = player_weights
    return weights
```

```python
decay_factor = 0.11
```

```python
def calculate_cum(column, decay_factor):
    for player in full['Player'].unique():
        temp = full[full['Player'] == player]
        for i,r in temp.iterrows():
            y = r['Season']
            total_num_seasons = max(temp['Season'])-min(temp['Season'])+1
            num_seasons_to_consider = total_num_seasons - (max(temp['Season']) - y)
            alphas = season_weights(decay_factor,num_seasons_to_consider) 
            Min_t = alphas[0] * r[column]
            if len(alphas) > 1:
                for c,d in temp.iterrows():
                    if d['Season'] < y:
                        Min_t += d[column] * alphas[y - d['Season']]
            full.loc[(full['Season']== y) & (full['Player']== player), column] = Min_t
```

```python
column_to_be_replaced_cum = ['MP','Min','Gls','Ast','PK','CrdY','CrdR','GlsxMin', 'AstxMin','G+AxMin']
new_column_names = {}
for col in column_to_be_replaced_cum:
    new_column_names[col] = 'cum_' + col
    calculate_cum(col,decay_factor)
full = full.rename(columns=new_column_names)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# create a dictionary of colors for each position
pos_colors = {'GK': 'red', 'DF': 'green', 'MF': 'blue', 'FW': 'orange', 'DFMF': 'darkgreen', 'FWMF': 'purple', 'DFFW': 'darkorange', 'MFFW': 'pink', 'MFDF': 'lightblue'}

# create the scatter plot with colored points
full.plot(kind='scatter', x='cum_Min', y='cum_Gls', c=full['Pos'].apply(lambda x: pos_colors[x]), colormap='viridis')

# set the plot title and axis labels
plt.title('Cumulative Minutes vs Goals by Position')
plt.xlabel('cum_Min')
plt.ylabel('cum_Gls')

# create the legend
handles = [plt.plot([], [], marker='o', linestyle='', color=color, label=pos)[0] for pos, color in pos_colors.items()]
plt.legend(handles=handles, title='Position', loc='upper left')

# display the plot
plt.show()
```
