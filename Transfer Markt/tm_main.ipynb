```python
contract_raw_2010 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2010.csv')
contract_raw_2011 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2011.csv')
contract_raw_2012 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2012.csv')
contract_raw_2013 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2013.csv')
contract_raw_2014 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2014.csv')
contract_raw_2015 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2015.csv')
contract_raw_2016 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2016.csv')
contract_raw_2017 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2017.csv')
contract_raw_2018 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2018.csv')
contract_raw_2019 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2019.csv')
contract_raw_2020 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2020.csv')
contract_raw_2021 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2021.csv')
contract_raw_2022 = pd.read_csv('https://raw.githubusercontent.com/edb-313/Football_players_trasfer_value/main/Transfer%20Markt/TM%20Data/benfica_contracts_2022.csv')
```

```python
contracts_2010 = contract_raw_2010.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2011 = contract_raw_2011.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2012 = contract_raw_2012.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2013 = contract_raw_2013.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2014 = contract_raw_2014.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2015 = contract_raw_2015.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2016 = contract_raw_2016.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2017 = contract_raw_2017.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2018 = contract_raw_2018.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2019 = contract_raw_2019.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2020 = contract_raw_2020.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2021 = contract_raw_2021.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
contracts_2022 = contract_raw_2022.drop(['comp_name', 'region', 'country','player_num','player_age','current_club','player_height_mtrs','player_url'], axis=1)
```

```python
all_seasons = pd.concat([contracts_2010, contracts_2011, contracts_2012, contracts_2013,contracts_2014,contracts_2015,contracts_2016,contracts_2017,contracts_2018,contracts_2019,contracts_2020,contracts_2021,contracts_2022], axis=0)
benfica_contracts = all_seasons.drop(['squad','joined_from'],axis=1)
benfica_contracts.dropna(inplace=True)
```
